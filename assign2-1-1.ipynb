```python
bias = 0
for i in range(epochs):
    j = 0
    for x in train_data:
        y_hat = w0*x[0] + w1*x[1] + w2*x[2] + bias
        
        if (y_hat >=0):
            act = 1
        else:
            act = 0
        err = op[j] - act
        
        del_w0 = alpha*x[0]*err
        del_w1 = alpha*x[1]*err
        del_w2 = alpha*x[2]*err
        
        w0 = w0 + del_w0
        w1 = w1 + del_w1
        w2 = w2 + del_w2
        
        j = j+1 
        print("epoch ", i+1, "error ", err)
        print(del_w0, del_w1, del_w2)

print(" final weight =")
print("w0 ", w0)
print("w1 ", w1)
print("w2", w2)
```
